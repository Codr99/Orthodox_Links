# Orthodox Christian Liturgical Texts — Afghan Languages

A curated list of HTML links to Orthodox Christian liturgical texts and closely related
Christian resources in the principal languages and dialects of Afghanistan.

---

## Research Context

Orthodox Christianity has **no indigenous community** in present-day Afghanistan.
The country is approximately 99% Muslim, and Christianity of any kind represents a tiny
and largely underground presence.  The only public Orthodox churches that have existed on
Afghan soil in modern times were Romanian military chapels built during the NATO mission
(e.g., Kandahar, ≈2009–2014), served in Romanian and English and since dismantled.

Consequently, **dedicated Orthodox liturgical texts in Afghan languages are extremely
scarce**, and some do not yet exist in digital form.  This notebook documents:

1. The few confirmed **Orthodox-specific** resources in Persian/Dari (the language closest
   to Tajiki, for which one translated Orthodox prayer book exists).
2. **Pan-Christian scripture translations** in Afghan languages — Scriptures are used in
   Orthodox worship and are the foundation of all Orthodox lectionary and hymnographic
   texts.
3. **Liturgically adjacent** Christian resources (Apostles' Creed, morning/evening
   prayers) in Dari and Pashto.
4. **Historical context** resources documenting the Church of the East's ancient
   presence on the Silk Road through what is now Afghanistan.

Each entry is tagged `[Orthodox]`, `[Pan-Christian]`, or `[Historical]` to make the
scope clear.

## 7. Classical Syriac — Historical / East Syriac Rite

The **Church of the East** (East Syriac Rite) was the dominant Christian presence across
Central Asia and into what is now Afghanistan during the 1st millennium.  Silk Road
trading communities brought the East Syriac liturgy as far as China (Nestorian Stele,
781 AD).  No living East Syriac community survives in Afghanistan today, but the
liturgical tradition is well-documented.

```python
import pandas as pd
from IPython.display import HTML

df = pd.DataFrame(links)

# Colour-code the tag column for quick scanning
tag_colours = {
    "Orthodox":      "background-color:#d4edda",  # green
    "Pan-Christian": "background-color:#fff3cd",  # yellow
    "Historical":    "background-color:#d1ecf1",  # blue
}

df["link"] = df["url"].apply(lambda u: f'<a href="{u}" target="_blank">{u}</a>')
df["tag_html"] = df["tag"].apply(
    lambda t: f'<span style="{tag_colours.get(t, "")};padding:2px 6px;border-radius:4px">{t}</span>'
)

HTML(
    df[["language", "tag_html", "source", "title", "link"]]
    .rename(columns={"tag_html": "scope"})
    .to_html(escape=False, index=False)
)
```

```python
# Summary counts by language and tag
summary = (
    df.groupby(["language", "tag"])
    .size()
    .reset_index(name="count")
    .sort_values(["language", "tag"])
)
HTML(summary.to_html(index=False))
```
