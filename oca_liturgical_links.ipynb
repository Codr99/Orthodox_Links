## 3. St. Tikhon's Monastery Press (STM Press)

The primary publisher of OCA-approved liturgical books in English.
STM Press editions of the service books are the standard for OCA parishes.

|---|---|

## 4. Saint Jonah Monastery — Orthodox Liturgical Texts

A comprehensive collection of full liturgical book texts in English, including
translations by Br. Isaac Lambertsen and Archimandrite Ephrem (Lash).

| Resource | URL |
|---|---|
| Orthodox Liturgical Texts and Resources (main hub) | https://saintjonah.org/orthodox-liturgical-texts-and-resources/ |
| Online Liturgical Resources index | https://saintjonah.org/online-liturgical-resources/ |
| All-Night Vigil | https://saintjonah.org/orthodox-liturgical-texts-and-resources/horologion/all-night-vigil |

## 5. All-Saints Eastern Orthodox Liturgical Resources (st-sergius.org)

Hosts full texts of the major liturgical books: Full Octoechos, General Menaion,
Full Menaion, Lenten Triodion, and Pentecostarion.

| Resource | URL |
|---|---|
| Main hub | https://st-sergius.org/ |

## 6. Ponomar Project

Academic project maintaining liturgical texts and typikon data; English-language section
includes digitized service texts.

| Resource | URL |
|---|---|
| Liturgical Texts in English | https://www.ponomar.net/english.html |

## 7. OrthodoxWiki

Reference articles on liturgical books with links to translations and publishers.

| Resource | URL |
|---|---|
| Liturgical Books article | https://orthodoxwiki.org/Liturgical_books |

## 8. Greek Orthodox Archdiocese Chapel (GOARCH)

The GOARCH online chapel provides English texts for all services.
While published by a different jurisdiction, many texts are shared across
jurisdictions and useful for cross-reference.

| Resource | URL |
|---|---|
| Liturgical Texts of the Orthodox Church | https://www.goarch.org/chapel/texts |

## 9. American Carpatho-Russian Orthodox Diocese (ACROD)

ACROD is in canonical communion with the Ecumenical Patriarchate but has close
historical ties to the OCA; their liturgical texts are largely compatible.

| Resource | URL |
|---|---|
| Liturgical Texts | https://www.acrod.org/prayercorner/liturgicaltexts/ |
