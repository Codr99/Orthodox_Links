## Research Context

Orthodox Christianity has a documented presence in China stretching back to **1685**,
when Russian Cossack prisoners of war brought their faith to Beijing and were eventually
granted a chapel (the Albazinian community).  The Russian Ecclesiastical Mission in
Beijing (founded 1715) produced the first systematic translations of Orthodox liturgical
books into Chinese, beginning in the 1830s.  Today the **Chinese Autonomous Orthodox
Church** (under the Moscow Patriarchate) and the **Orthodox Metropolitanate of Hong Kong
and Southeast Asia** (under the Ecumenical Patriarchate) represent the two main
jurisdictions.

### Linguistic Situation

| Chinese variety | Speakers | Script convention | Orthodox liturgical use |
|---|---|---|---|
| **Mandarin / Putonghua** (普通话) | ~920 M (70 %) | Simplified (PRC) / Traditional (TW, HK, diaspora) | **Primary** — all known Chinese Orthodox liturgical texts |
| **Wu** (incl. Shanghainese 上海话) | ~80 M (6 %) | Simplified | No confirmed Orthodox-specific texts |
| **Yue / Cantonese** (粤语) | ~68 M (5 %) | Traditional | One confirmed inculturation project (HK, 2022); Lord's Prayer only |
| **Min** (incl. Hokkien/Fujianese 闽南语) | ~60 M (5 %) | Simplified/Traditional | No confirmed Orthodox-specific texts |
| **Hakka** (客家话) | ~48 M (4 %) | Simplified | No confirmed Orthodox-specific texts |
| **Gan** (赣语) | ~31 M (2 %) | Simplified | No confirmed Orthodox-specific texts |
| **Xiang** (湘语) | ~38 M (3 %) | Simplified | No confirmed Orthodox-specific texts |

> **Note on script vs. dialect:** Simplified and Traditional characters are two written
> forms of the same *Standard Chinese / Mandarin* — they are not separate dialects.  The
> distinction is politically and regionally significant (PRC uses Simplified; Taiwan,
> Hong Kong, and the overseas Chinese diaspora largely use Traditional), and it affects
> which version of a liturgical text a parish uses.  Both are catalogued below.

### Key Three-Character Script Periods

The 19th-century Russian Ecclesiastical Mission translated liturgical books into
**Classical Chinese** (文言文 — the literary language used before vernacular Mandarin
became standard).  Modern Orthodox translations use **Modern Standard Chinese**
(现代汉语).  Both corpora are included below.

## 7. Historical Context — Tang Dynasty Jingjiao (景教) / Church of the East

The **Church of the East** (*Jingjiao* — "Luminous Religion") arrived in China in 635 AD
under the Persian monk Alopen and flourished for ≈200 years during the Tang Dynasty.
It was a distinct tradition (not Byzantine Orthodox), using Classical Syriac and a
Syriac-influenced Chinese liturgical language.  The Xi'an Stele (781 AD) — the
world-famous bilingual Chinese–Syriac inscription — is the primary documentary evidence
of this church's Chinese liturgy.
