# Orthodox Christian Liturgical Texts — Indian Languages

A curated list of HTML links to Orthodox Christian liturgical texts in the most common
Indian (India) languages, including Malayalam, Tamil, Hindi, Kannada, Telugu, and other
regional languages.

## Background

The primary Orthodox Christian presence in India is the **Oriental Orthodox** tradition,
centred on two sister churches rooted in Kerala:

| Church | Jurisdiction | Primary liturgical language |
|---|---|---|
| **Malankara Orthodox Syrian Church (MOSC)** — Indian Orthodox Church (IOC) | Autocephalous (Catholicos of the East) | Malayalam + West Syriac |
| **Jacobite Syrian Christian Church** | Maphrianate of the Syriac Orthodox Church of Antioch | Malayalam + Classical Syriac |

Both churches use the **West Syriac Rite / Liturgy of Saint James**.  Malayalam is the
overwhelming vernacular for both; additional Indian languages (Tamil, Kannada, Telugu,
Hindi, Konkani, Marathi) are used in mission regions and diaspora contexts.

Resources below are organised first by primary language, then by institution.
