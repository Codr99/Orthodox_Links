# Orthodox Christian Liturgical Texts — Southeast Asian Languages

A curated list of HTML links to Orthodox Christian liturgical texts in the principal
languages of Southeast Asia: Vietnam, Indonesia, Myanmar, Laos, Thailand, the Philippines,
Cambodia, Malaysia, and Timor-Leste.

---

## Research Context

Orthodox Christianity is a minority faith across Southeast Asia.  Three overlapping
jurisdictions serve the region:

| Jurisdiction | Mother Church | Coverage |
|---|---|---|
| **Patriarchal Exarchate of Southeast Asia** (PESEA, est. 2018) | Russian Orthodox Church (Moscow Patriarchate) | Indonesia, Malaysia, Singapore, Philippines, Vietnam, Thailand, Cambodia, Laos, Myanmar |
| **Orthodox Metropolitanate of Hong Kong and Southeast Asia** (OMHKSEA, est. 1996) | Ecumenical Patriarchate of Constantinople | Philippines, Thailand, Vietnam, Laos, Cambodia, Myanmar, HK, Taiwan |
| **Eastern Orthodox Metropolitanate of Singapore and South Asia** (est. 2008) | Ecumenical Patriarchate of Constantinople | Singapore, Indonesia, Malaysia, Brunei, Timor-Leste |

A fourth independent body, **Gereja Orthodox Indonesia (GOI)**, is affiliated with the
Genuine Greek Orthodox Church (GGOC) and produces the largest corpus of Indonesian-language
liturgical texts.

### Language Coverage Overview

| Language | Approx. speakers | Orthodox liturgical resources |
|---|---|---|
| **Indonesian** (Bahasa Indonesia) | ~270 M (+ Malay) | **Extensive** — 40+ free PDFs, full Paschal cycle, Akathists |
| **Javanese** | ~82 M | Translation underway; no confirmed online texts |
| **Filipino / Tagalog** | ~45 M | liturhiya.info; physical prayer book |
| **Cebuano** | ~20 M | liturhiya.info; published Divine Liturgy & prayer book |
| **Thai** | ~60 M | Published liturgical collection & Presanctified Liturgy; no public download link found |
| **Vietnamese** | ~95 M | YouTube catechesis; grassroots prayer translations |
| **Burmese** (Myanmar) | ~33 M | No confirmed Orthodox-specific texts |
| **Khmer** (Cambodia) | ~16 M | No confirmed Orthodox-specific texts |
| **Lao** | ~7 M | No confirmed Orthodox-specific texts |
| **Malay** (Bahasa Melayu) | ~33 M | No confirmed Orthodox-specific texts |
| **Tetum / Timorese** | ~0.8 M | One confirmed Divine Liturgy PDF (OMSGSA) |
| **Waray** | ~3 M | Hours & Typika translated; physical edition only |

> **Note:** The Lord's Prayer (*Pater Noster*) — recited at every Orthodox Divine Liturgy —
> exists in nearly all Southeast Asian languages and is included below as a shared
> pan-Christian resource useful for Orthodox worship.

## 4. Vietnamese — Tiếng Việt

Orthodox Christianity arrived in Vietnam via Russian diplomatic and military contact.
The Philippine-Vietnamese Diocese (ROC) includes Vietnam, and converts are increasing;
however, published Vietnamese Orthodox liturgical texts remain limited.  Translation work
is primarily grassroots and ongoing.
